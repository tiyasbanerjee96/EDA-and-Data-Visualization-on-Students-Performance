### Data Visualization

```python
sns.pairplot(d)
```

```python
sns.pairplot(d,diag_kind='kde',kind='reg')
```

```python
sns.relplot(x='math score',y='reading score',hue='gender',data=d)
```

```python
sns.distplot(d['math score'])
```

```python
sns.catplot(x='math score',y='gender',kind='box',data=d)
```

### To create Heatmap for checking correlation


```python
sns.heatmap(d.corr())
```

```python
sns.lineplot(data= d, x='gender',y='math score', color='blue')
```

```python
sns.stripplot(x = d["gender"],y = d["math score"])
plt.title("gender VS math score")
```

```python
sns.swarmplot(x = d["gender"], y = d["math score"],hue = d["gender"],s = 4)
plt.title("gender VS math score")
```
